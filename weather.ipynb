```python
file_path = "/content/drive/MyDrive/weatherHistory.csv"
```

```python
import numpy as np
import pandas as pd
```

```python
data=pd.read_csv(file_path)
```

```python
data = {
    "Day": ["Mon", "Tue", "Wed", "Thu", "Fri"],
    "Temperature": [30, 32, 31, 29, 28],
    "Humidity": [65, 70, 68, 72, 75],
    "WindSpeed": [10, 12, 9, 11, 8]
}

weather = pd.DataFrame(data)

print(weather)
```

```python
import matplotlib.pyplot as plt

plt.plot(weather["Day"], weather["Temperature"])
plt.xlabel("Day")
plt.ylabel("Temperature")
plt.title("Weather Temperature Data")
plt.show()
```
